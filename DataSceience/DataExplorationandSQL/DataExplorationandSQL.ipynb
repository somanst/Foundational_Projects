# 3rd Set of Questions on SQL - **using sqlite3 library!** **(40 Pts)**

* **Dataset Path: "Data/annots.csv"**

Here you are given a file: annots.csv, with the protein - gene ontology (GO) term associations/annotations. Gene ontology is a framework, representing the properties of genes and proteins. This is basically a Graph - more spesifically a directed acyclic graph (DAG), representing a level-ordered hierarchy with three subgraphs: MF, CC, and BP. Here, **each** node in the graph, which we term as **"GO Term" corresponds to a function of a protein.**

* If protein is annotated with a GO Term, then the protein has the function represented by this GO Term. 
* A protein can be and most of the time is annotated with multiple GO terms, and vice versa. 

In the annots.csv file, we have the annotation data, where each row corresponds to a protein - GO term annotation with the following columns: ProteinID,ProtSeq, GO_ID, Sub_Graph, Level. ProteinID is the specifier of the protein and the ProtSeq is the aminoacid sequence of this protein. GO_ID, Sub_Graph, and Level are properties of a GO Term. 

**Now, we want to store this data in a relational database, with the following table schemas:**

## Table 1: GoTerms:

| Column   | Type    | Key        | Description                |
|----------|---------|------------|----------------------------|
| GoID     | TEXT    | PRIMARY KEY| Gene Ontology Identifier   |
| SubGraph | TEXT    |            | Subgraph Type (BP, MF, CC) |
| Level    | INTEGER |            | Ontology level             |


## Table 2: Proteins

| Column     | Type | Key        | Description                |
|------------|------|------------|----------------------------|
| ProteinID  | TEXT | PRIMARY KEY| Unique Protein Identifier  |
| ProtSeq  | TEXT | | Amino Acid Sequence |

## Table 3: Annotations

| Column    | Type | Key                       | Description               |
|-----------|------|---------------------------|---------------------------|
| GoID      | TEXT | PRIMARY KEY, FOREIGN KEY  | Gene Ontology Identifier  |
| ProteinID | TEXT | PRIMARY KEY, FOREIGN KEY  | Unique Protein Identifier |


## Q1. Read the csv file and create this DB  **(10 Pts)**

```python
# read file into db and create a db using sqlite3
annoDf = pd.read_csv("Data/annots.csv")

conn = sqlite3.connect("AnnotsDB.db")
cursor = conn.cursor()

#copying the required part of the main dataframe for each table.
GoTermsData = annoDf[["GO_ID", "Sub_Graph", "Level"]].copy()
ProteinsData = annoDf[["ProteinID", "ProtSeq"]].copy()
AnnotsData = annoDf[["GO_ID", "ProteinID"]].copy()


#Schemas of databases
GoTermSchema="""
             CREATE TABLE IF NOT EXISTS GoTerms(
             "GoID" TEXT,
             SubGraph TEXT,
             Level INTEGER,
             PRIMARY KEY (GoID))
             """

ProteinSchema="""
              CREATE TABLE IF NOT EXISTS Proteins(
              ProteinID TEXT,
              ProtSeq TEXT,
              PRIMARY KEY (ProteinID))
              """

AnnotSchema="""
            CREATE TABLE IF NOT EXISTS Annotations( 
            "GoID" TEXT,
            ProteinID TEXT,
            PRIMARY KEY (GoID, ProteinID),
            FOREIGN KEY (GoID) REFERENCES GoTerms(GoID),
            FOREIGN KEY (ProteinID) REFERENCES Proteins(ProteinID))
            """

#Executing schemas
cursor.execute(GoTermSchema)
cursor.execute(ProteinSchema)
cursor.execute(AnnotSchema)

#Renaming columns so that data can be transfered smoothly
GoTermsData.rename(columns={"GO_ID" : "GoID"}, inplace=True)
AnnotsData.rename(columns={"GO_ID" : "GoID"}, inplace=True)

#Transferring data
GoTermsData.to_sql("GoTerms", conn, if_exists='replace', index=False)
ProteinsData.to_sql("Proteins", conn, if_exists='replace', index=False)
AnnotsData.to_sql("Annotations", conn, if_exists='replace', index=False)
```

## Q3. Print the averge number of annotations a GO terms has for each level in each subgraph (MF, CC, and BP) **(10 Pts)**

```python
# Print the avg # of annots

#Drop so we can create a table more than once
dropQuery = """
            DROP TABLE IF EXISTS GroupedTable;
            """ 
#Table to make calculations on
query = """
        CREATE Table GroupedTable AS
        SELECT g.Sub_Graph, Level, g.GoID, COUNT(a.GoID) AS AnnotationCount
        FROM Annotations a, GoTerms g
        WHERE a.GoID = g.GoID
        GROUP BY g.Sub_Graph, g.Level, g.GoID
        """

cursor.execute(dropQuery)
cursor.execute(query)

#Calculations on table
averageQuery = """
               SELECT Sub_Graph, Level, AVG(AnnotationCount) AS AnnotationAvg
               FROM GroupedTable
               GROUP BY Sub_Graph, Level
               """
cursor.execute(averageQuery)
cursor.fetchall()
```

We can see that the higher the level, the much less the average number of annotations for each GOterm becomes across all
subgraphs, while the lower levels have way much annotations, being MF and BP having more annotations than CC subgraph. and the
higher we go with levels, the less the average.

## Q4. Identify the ProteinID with the longest ProtSeq within the Proteins table. Modify its ProtSeq to half its original length by trimming the sequence, update the ProteinID to "trimmed", and display the updated ProteinID and ProtSeq with its GO Term annotations. **(10 Pts)**

```python
# find and trim the longest sequence, displaying the updated ProtID and ProtSeq, and its annotated GO Terms

#Drop so we can create a table more than once
dropQuery = """
            DROP TABLE IF EXISTS LongestSeq;
            """ 

#Create a table that has longest protein sequence.
query = """
        CREATE Table LongestSeq AS
        SELECT p.ProteinID, ProtSeq, GoID
        FROM Proteins p, Annotations a
        WHERE p.ProteinID = a.ProteinID AND
              LENGTH(ProtSeq) = (
                  SELECT MAX(LENGTH(ProtSeq))
                  FROM Proteins)
        """

cursor.execute(dropQuery)
cursor.execute(query)

#Updates our table data, adding "trimmed" to the proteinID
updateQuery = """
              UPDATE LongestSeq 
              SET ProtSeq = SUBSTR(ProtSeq, 1, LENGTH(ProtSeq) / 2),
                  ProteinID = 'trimmed_' || ProteinID;
              """

cursor.execute(updateQuery)
cursor.execute("""SELECT * FROM LongestSeq""")
cursor.fetchall()
```

This give us the GoTerms annotated with the longest protseq, displaying updated data of Protein table, adding trimmed to 
ProteinID and cutting ProtSeq in half.
